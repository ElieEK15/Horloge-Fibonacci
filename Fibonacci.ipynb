```python
from p5 import *
from datetime import *
from random import *

# Définition des segments qui composent les chiffres
segments = [
    [(50, 20), (110, 20), (100, 30), (60, 30)],  # Segment A
    [(110, 30), (120, 40), (120, 80), (110, 90)],  # Segment B
    [(110, 90), (120, 100), (120, 140), (110, 150)],  # Segment C
    [(50, 140), (110, 140), (100, 150), (60, 150)],  # Segment D
    [(40, 90), (50, 100), (50, 140), (40, 130)],  # Segment E
    [(40, 30), (50, 40), (50, 80), (40, 70)],  # Segment F
    [(50, 80), (110, 80), (100, 90), (60, 90)],  # Segment G
]

# Association des chiffres avec les segments à allumer
numbers = {
    0: [0, 1, 2, 3, 4, 5],
    1: [1, 2],
    2: [0, 1, 6, 4, 3],
    3: [0, 1, 6, 2, 3],
    4: [5, 6, 1, 2],
    5: [0, 5, 6, 2, 3],
    6: [0, 5, 6, 2, 3, 4],
    7: [0, 1, 2],
    8: [0, 1, 2, 3, 4, 5, 6],
    9: [0, 1, 2, 3, 5, 6],
}

def setup():
    """Initialisation de la fenêtre d'affichage."""
    createCanvas(2000, 200)
    background(0)

def afficher_chiffre(n, dist,couleur):
    """Affiche un chiffre à une position donnée."""
    if n not in numbers:
        return

    fill(couleur)
    
    for i in numbers[n]:  
        shape = segments[i]
        beginShape()
        for px, py in shape:
            vertex(px + dist, py)
        endShape(CLOSE)

def draw():
    """Dessine les chiffres de 0 à 9 en ligne."""
    background(0)
    
    for i in range(10):
        afficher_chiffre(i, i * 200,(255))

run()
```

```python
def setup():
    """
    Initialise la fenêtre avec un fond noir et un taux de rafraîchissement de 1 image par seconde.
    """
    createCanvas(730, 200)  # Taille de la fenêtre adaptée à l'affichage de l'horloge
    background(0)  # Fond noir
    frameRate(1)  # Mise à jour chaque seconde pour afficher l'heure en temps réel

def draw():
    """
    Met à jour l'affichage de l'horloge à chaque seconde.
    """
    background(0)  # Efface l'écran à chaque rafraîchissement

    # Récupère l'heure actuelle
    now = datetime.now()
    heures = now.hour
    minutes = now.minute
    secondes = now.second

    # Affichage des chiffres de l'heure
    # Les dizaines et unités de chaque composante (heures, minutes, secondes) sont affichées séparément
    afficher_chiffre(heures // 10, 20,(0,255,0))    # Chiffre des dizaines des heures
    afficher_chiffre(heures % 10, 120,(0,255,0))    # Chiffre des unités des heures
    afficher_chiffre(minutes // 10, 240,(0,255,0))  # Chiffre des dizaines des minutes
    afficher_chiffre(minutes % 10, 340,(0,255,0))   # Chiffre des unités des minutes
    afficher_chiffre(secondes // 10, 460,(0,255,0)) # Chiffre des dizaines des secondes
    afficher_chiffre(secondes % 10, 560,(0,255,0))  # Chiffre des unités des secondes

    # Affichage des deux-points (:) entre les heures, minutes et secondes
    fill(0, 255, 0)  # Vert fluo pour correspondre au style des chiffres

    # Affiche deux points pour séparer les heures, minutes et secondes
    for x in [260, 485]:  # Coordonnées X des séparateurs
        ellipse(x, 85, 10, 10)  # Point supérieur
        ellipse(x, 145, 10, 10) # Point inférieur

run()
```

```python
# Initialisation du compte à rebours
heures = int(input("Entrez les heures : "))
minutes = int(input("Entrez les minutes : "))
secondes = int(input("Entrez les secondes : "))

countdown = timedelta(hours=heures, minutes=minutes, seconds=secondes)

def setup():
    createCanvas(730, 600)
    frameRate(1)  # 1 FPS pour le compte à rebours

def draw():
    global countdown
    background(0)

    if countdown.total_seconds() > 0:
        # Convertir en heures, minutes, secondes
        h = countdown.seconds // 3600
        m = (countdown.seconds % 3600) // 60
        s = countdown.seconds % 60

        # Afficher les chiffres
        afficher_chiffre(h // 10, 20, (255, 0, 0))    
        afficher_chiffre(h % 10, 120, (255, 0, 0))    
        afficher_chiffre(m // 10, 240, (255, 0, 0))  
        afficher_chiffre(m % 10, 340, (255, 0, 0))   
        afficher_chiffre(s // 10, 460, (255, 0, 0)) 
        afficher_chiffre(s % 10, 560, (255, 0, 0))  

        # Réduire le temps restant
        countdown -= timedelta(seconds=1)

    else:
        frameRate(30)  # 30 FPS pour les confettis
        afficher_confettis()

def afficher_confettis():
    """ Affiche une explosion de confettis aléatoires """
    for _ in range(250):
        x = randint(0, width)
        y = randint(0, height)
        fill(randint(0, 255), randint(0, 255), randint(0, 255))
        ellipse(x, y, randint(1,10), randint(1, 10))

run()
```

```python
# Définition des carrés de l'horloge de Fibonacci (x, y, taille)
squares = [
    (310, 100, 295),  # 5
    (130, 220, 175),  # 3
    (130, 100, 115),  # 2
    (250, 160, 55),   # 1
    (250, 100, 55)    # 1
]

values = [5, 3, 2, 1, 1]  # Correspondance des tailles

def setup():
    createCanvas(750, 500)
    frameRate(1)

def methode(duree):
    """Convertit une valeur (heures ou minutes) en activation des carrés."""
    L = [0] * 5
    for i in range(5):
        if duree >= values[i]:
            duree -= values[i]
            L[i] = 1
    return L

def draw():
    now = datetime.now()
    background(48, 25, 100) if now.hour >= 12 else background(249, 231, 64) #Le fond est jaune lorsque l'heure est AM et est mauve lorsque l'heure est PM

    hour_list = methode(now.hour % 12)
    minute_list = methode(now.minute // 5)

    for i in range(5):
        if hour_list[i] and minute_list[i]:
            fill(0, 0, 255)  # Bleu (heures + minutes)
        elif hour_list[i]:
            fill(255, 0, 0)  # Rouge (heures)
        elif minute_list[i]:
            fill(0, 255, 0)  # Vert (minutes)
        else:
            fill(255)  # Blanc (vide)

        square(*squares[i])  # Dessine le carré

run()
```

```python
def setup():
    createCanvas(400, 300)  # Taille de la fenêtre
    frameRate(1)  # Mettre à jour une fois par seconde

def draw():
    background(255)
    now = datetime.now()
    
    heures = bin(now.hour)[2:].zfill(6)    # Convertit en binaire (6 bits)
    minutes = bin(now.minute)[2:].zfill(6) # Convertit en binaire (6 bits)
    secondes = bin(now.second)[2:].zfill(6) # Convertit en binaire (6 bits)

    afficher_binaire(heures, 50, "H")
    afficher_binaire(minutes, 150, "M")
    afficher_binaire(secondes, 250, "S")

def afficher_binaire(bits, y, label):
    """Affiche une rangée de cercles pour représenter un nombre binaire."""
    fill(0)  
    textSize(20)
    text(label, 20, y + 10)  # Affiche H, M ou S devant la ligne

    for i in range(6):
        x = 80 + i * 50  # Espacement horizontal
        if bits[i] == "1":
            fill(0, 255, 255)  # Allumé (Cyan)
        else:
            fill(200)  # Éteint (gris clair)
        ellipse(x, y, 30, 30)  # Dessine le cercle

run()
```
